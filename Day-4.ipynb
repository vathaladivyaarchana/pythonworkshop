```python
#prime numbers b/w range
def Prime(n):
    count=0
    for i in range(1,n+1):
        if(n%i==0):
            count+=1
    if count==2:
        return True
    else: 
        return False

s=int(input())
e=int(input())
Prime_tuple=tuple([i for i in range(s,e+1) if Prime(i)])
Prime_tuple
#len(Prime_list)
```

## modes

- for write the data into the file  we use 'w+"
- for adding the data into the file we use "a+"

```python
#f=open("pace.txt","x")
#open()
#syntax: open(filename,mode)
f=open("pace.txt",'x')
f.close()
```

```python
f=open("pace.txt",'r+')
data=f.read()
print(data)
f.close()
```

```python
f=open("pace.txt",'a+')
data="welcome to cse second year"
f.append(data)
print("data write into the file is successfully...")
f.close()
```

```python
f=open("pace.txt",'a+')
data="cse-A section welcome to python programming"
f.write(data)
f.close()
```

```python
f=open("pace.txt",'w+')
f.write("hello")
data=f.read()
print(data)
f.close()
```

```python
f=open("pace.txt",'r')
data=f.readlines()#read(10)-->read 10 characters
print(data)
f.close()
```

```python
f=open("pace.txt",'a')
data='\nabc\ncde\ndef'
f.write(data)
f.close()

```

```python
# wrire data
with open("pace.txt",'w') as f:
    data="bhavani:23,15"
    f.write(data)
```

```python
#read data
with open("pace.txt",'r') as f:
    data=f.read()
    print(data)
```

```python
with open("pace.txt",'a') as f:
    data='\nabc :36,96'
    f.write(data)
```

```python
with open("pace.txt",'a') as f:
    print(f.tell())
    f.seek(10)
```

```python
with open("pace.txt",'a') as f:
    print(f.tell())
    print(f.read())
    f.seek(10)
    print(f.read)
    f.write("\nabc: 96,36")
    
```
